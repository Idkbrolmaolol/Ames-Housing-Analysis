```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/raw/AmesHousing.csv')

def clean_data(raw_df):
    pd.set_option('future.no_silent_downcasting', True)
    df_clean = raw_df.copy()
    df_clean['MS SubClass'] = df_clean['MS SubClass'].astype('object')
    df_clean['Overall Qual'] = df_clean['Overall Qual'].astype('object')
    df_clean = df_clean.drop_duplicates()
    num_cols = df_clean.select_dtypes(include=[np.number]).columns
    for col in num_cols:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())
    cat_cols = df_clean.select_dtypes(include=['object']).columns
    for col in cat_cols:
        df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
    target = 'SalePrice'
    q99 = df_clean[target].quantile(0.99)
    df_clean[target] = np.where(df_clean[target] > q99, q99, df_clean[target])
    return df_clean

df_phase1 = clean_data(df)
df_phase1.to_csv('data/cleaned/cleaned_data.csv', index=False)
print("Phase 1 complete!")
```
