```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/cleaned/featured_data.csv')

plt.figure(figsize=(8, 5))
sns.histplot(df['SalePrice'], color='blue', kde=True)
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Gr Liv Area', y='SalePrice', hue='Overall Qual', data=df)
plt.show()

prices = df['SalePrice'].values
my_mean = np.sum(prices) / len(prices)
my_std = np.sqrt(np.sum((prices - my_mean)**2) / len(prices))

print(f"Average: {my_mean}")
print(f"Spread: {my_std}")
```
